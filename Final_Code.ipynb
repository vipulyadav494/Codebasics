# Q2. List down the top 3 District based compound annual growth rate (CAGR) of visitors between (2016-2019)?<br>(Insight: District that are declining)


CAGR Formula=(V-final/V-begin)^(1/t)<br>
V-final= Final Value<br>
V-begin= Begining Value<br>
t= Time

If Begining Value are Zero here we will calculat AAGR for those values.<br>
AAGR Formula= (V Final-V Begining)/Time

````python
df_total=pd.concat([df_domastic,df_foreign])
df_total
````

````python
def CAGR(district):
    Temp=df_total[df_total['district']==district].groupby('year')['visitors'].sum()
    X=list(pd.DataFrame(Temp)['visitors'])
    t=2019-2016+1
    V_F=X[0]         # 2016
    V_B=X[-1]        #2019
    if V_B!=0:
        CAGR=(V_F/V_B)**(1/t)
        return district,(CAGR-1)*100,0
    else:
        AAGR=(V_F-V_B)/t
        return district,(AAGR)*100,1
````

````python
Districts=list(set(df_total['district']))
````

````python
df_CAGR=pd.DataFrame([CAGR(i) for i in Districts],columns=['district',"value","CAGR/AAGR"])
df_CAGR
````

````python
#CAGR=0, AAGR=1
df_CAGR[df_CAGR["CAGR/AAGR"]==0].sort_values('value',ascending=False).head(3)
````

````python
def calculate_ratio(list1, list2):
    """
    This function takes two lists as input and returns a new list with the element-wise ratio of the input lists.
    """
    result = []
    for i in range(len(list1)):
        # Avoid division by zero by checking if the denominator is zero
        if list2[i] == 0:
            ratio = float('inf')
        else:
            ratio = round(list1[i] / list2[i],2)
        result.append(ratio)
    return result
````

# Q7. What will be the projected number of domestic and foreign tourists in Hyderabad in 2025 based on the growth rate from previous years?<br>(Insight: Better estimate of incoming tourists count so that government can plan the infrastructure better)



Formula:<br>
`Growth Rate = ((Current Year Visitors - Previous Year Visitors) / Previous Year Visitors) * 100`

````python
def VisitorPred(Visitors,Years,TargetYear):
  Count=len(Years)-1
  GrothRate=[]
  Duration=[]
  for i in range(Count):
    GR=((Visitors[i+1]-Visitors[i])/Visitors[i])*100
    Duration.append(str(Years[i])+'-'+str(Years[i+1]))
    GrothRate.append(GR)
  AvgGrothRate=round(sum(GrothRate)/len(GrothRate),ndigits=2)
  Count=TargetYear-Years[-1]
  LastYear=Years[-1]
  LastYearVisitor=Visitors[-1]
  PredVisitors=[]
  PVYear=[]
  for i in range(Count):
    Projection=LastYearVisitor+(LastYearVisitor*AvgGrothRate)/100
    LastYearVisitor=Projection
    LastYear=LastYear+1
    PredVisitors.append(int(Projection))
    PVYear.append(LastYear)
  return PVYear,PredVisitors
````

````python
df_Dom_HYD=df_domastic[df_domastic['district']=='Hyderabad']
df_For_HYD=df_foreign[df_foreign['district']=='Hyderabad']
````

````python
VisitorsDomastic=list(df_Dom_HYD.groupby('year')['visitors'].sum().values)
VisitorsForeign=list(df_For_HYD.groupby('year')['visitors'].sum().values)
Years=list(df_total.groupby('year')['visitors'].sum().index)
````

````python
temp=VisitorPred(VisitorsDomastic,Years,2025)
PVYear=temp[0]
PredVisitorsDomastic=temp[1]

temp=VisitorPred(VisitorsForeign,Years,2025)
PredVisitorsForeign=temp[1]
````
